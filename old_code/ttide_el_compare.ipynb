## Load imports and the drag ncfile

```python
from __future__ import division
import matplotlib as mpl
import scipy as sp
from datatools import *
from gridtools import *
from plottools import *
from projtools import *
import matplotlib.tri as mplt
import matplotlib.pyplot as plt
#from mpl_toolkits.basemap import Basemap
import os as os
import sys
import pandas as pd
from matplotlib.collections import LineCollection as LC
```

```python
# Define names and types of data
name_orig='kit4_kelp_nodrag'
name_change='kit4_kelp_20m_drag_0.018'
grid='kit4_kelp'
datatype='2d'
regionname='kit4_kelp_tight2_kelpfield'


### load the .nc file #####
data = loadnc('runs/'+grid+'/'+name_change+'/output/',singlename=grid + '_0001.nc')
print 'done load'
data = ncdatasort(data)
print 'done sort'

cages=loadcage('runs/'+grid+'/' +name_change+ '/input/' +grid+ '_cage.dat')
if cages!=None:
    tmparray=[list(zip(data['nodell'][data['nv'][i,[0,1,2,0]],0],data['nodell'][data['nv'][i,[0,1,2,0]],1])) for i in cages ]
    color='k'
    lw=.2
    ls='solid'

region=regions('kit4_kelp_tight2_kelpfield')
nidx=get_nodes(data,region)

savepath='figures/png/' + grid + '_' + datatype + '/ttide_el_reldiff/' + name_orig + '_' + name_change + '/'
if not os.path.exists(savepath): os.makedirs(savepath)
```

## Load the data from ttide for the elevation

```python
elo=np.load('data/ttide/kit4_kelp_kit4_kelp_nodrag_2d_el_all.npy')
elo=elo[()]
elc=np.load('data/ttide/kit4_kelp_kit4_kelp_20m_drag_0.018_2d_el_all.npy')
elc=elc[()]
```

## Find the mean and max difference in the constituents over the area.

```python
condiff=elc['tidecon'][nidx,:,0]-elo['tidecon'][nidx,:,0]
condiff_mean=np.mean(np.fabs(condiff),axis=0)
condiff_max=np.max(np.fabs(condiff),axis=0)

df=pd.DataFrame(np.vstack([condiff_mean,condiff_max]).T,elo['nameu'],['Mean Amp. Diff.','Max Amp. Diff.'])
```

## Sort by size of the mean amplitude and display.

```python
print df.sort('Mean Amp. Diff.',ascending=False)
```

## Sort by size of the max amplitude and display.

```python
print df.sort('Max Amp. Diff.',ascending=False)
```

## Plot the amplitude relative difference for M2.

```python
#add 1mm to avoid division by zero.
conname='M2  '
conidx=np.argwhere(elo['nameu']==conname)
reldiff=100*((elc['tidecon'][:,conidx,0]-elo['tidecon'][:,conidx,0])/(elo['tidecon'][:,conidx,0]+.001)).flatten()

f=plt.figure(figsize=(12,12))
ax=f.add_axes([.125,.1,.775,.8])
triax=ax.tripcolor(data['trigrid'],reldiff,vmin=-.15,vmax=.15,cmap=mpl.cm.seismic)
prettyplot_ll(ax,setregion=region,cblabel=r'Amp. Relative Difference (%)',cb=triax)
plotcoast(ax,filename='pacific.nc',color='.75',fill=True)
if cages!=None:   
    lseg_t=LC(tmparray,linewidths = lw,linestyles=ls,color=color)
    ax.add_collection(lseg_t) 
    
f.savefig(savepath + grid + '_' + regionname+'_el_compare_reldiff_'+conname.strip()+'.png')
```
