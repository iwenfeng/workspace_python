```python
sys.path.append('/home/moflaher/Desktop/workspace_python/ttide_py/ttide/')
from t_tide import t_tide
from t_predic import t_predic
```

```python
data = loadnc('/media/moflaher/My Book/kit4_runs/kit4_45days_3/output/')
data= ncdatasort(data)
data['time']=data['time']+55545
grid='kit4'
datatype='2d'
regionname='gilisland'
region=regions(regionname)
elements=get_elements(data,region)
elementmarker=elements[::10000]
starttime=383
endtime=1081
```

```python
cnt=0
mypredicu=np.empty([len(elementmarker),len(data['time'][starttime:endtime])])
mypredicv=np.empty([len(elementmarker),len(data['time'][starttime:endtime])])
for i in elementmarker:
    [nameu, freq, tidecon, xout]=t_tide(data['ua'][starttime:endtime,i]+1j*data['va'][starttime:endtime,i],stime=data['time'][starttime],lat=data['uvnodell'][i,1],output=False)
    mypredicu[cnt,:]=np.real(t_predic(data['time'][starttime:endtime],nameu,freq,tidecon)).flatten()
    mypredicv[cnt,:]=np.imag(t_predic(data['time'][starttime:endtime],nameu,freq,tidecon)).flatten()
    cnt=cnt+1
```

```python
print len(elementmarker)
for i in range(0,len(elementmarker)):
    f, (ax1, ax2, ax3, ax4) = plt.subplots(4, sharex=True, sharey=False)
    ax1.plot(data['time'][starttime:endtime],mypredicu[i,:],'r',label='ttide')
    ax1.plot(data['time'][starttime:endtime],data['ua'][starttime:endtime,elementmarker[i]],'b',label='model')   
    ax1.grid()
    ax1.set_title('Currents at: %d' % (elementmarker[i]))
    ax1.set_ylabel(r'ua ($\frac{m}{s}$)')
    handles, labels = ax1.get_legend_handles_labels()
    ax1.legend(handles[::-1], labels[::-1])
    
    ax2.plot(data['time'][starttime:endtime],mypredicv[i,:],'r',label='ttide')
    ax2.plot(data['time'][starttime:endtime],data['va'][starttime:endtime,elementmarker[i]],'b',label='model')
    ax2.grid()
    ax2.set_ylabel(r'va ($\frac{m}{s}$)')
    handles, labels = ax2.get_legend_handles_labels()
    ax2.legend(handles[::-1], labels[::-1])
    
    ax3.plot(data['time'][starttime:endtime],data['ua'][starttime:endtime,elementmarker[i]]-mypredicu[i,:],'r',label='ua residual')
    ax3.plot(data['time'][starttime:endtime],data['va'][starttime:endtime,elementmarker[i]]-mypredicv[i,:],'b',label='va residual')
    ax3.grid()
    ax3.set_ylabel(r'Residual ($\frac{m}{s}$)')
    handles, labels = ax3.get_legend_handles_labels()
    ax3.legend(handles[::-1], labels[::-1])
    

    ax4.plot(data['time'][starttime:endtime],np.gradient((data['zeta'][starttime:endtime,data['nv'][elementmarker[i],0]]+data['zeta'][starttime:endtime,data['nv'][elementmarker[i],1]]+data['zeta'][starttime:endtime,data['nv'][elementmarker[i],2]])/3),'y',alpha=.5,label='Gradient zeta')
    q=ax4.quiver(data['time'][starttime:endtime],0,data['ua'][starttime:endtime,elementmarker[i]]-mypredicu[i,:],data['va'][starttime:endtime,elementmarker[i]]-mypredicv[i,:],scale = 1,headlength=3,headaxislength=2,width=0.002,alpha=0.75,label='Residual Direction')
    qk = plt.quiverkey(q,0.2, 0.05, 0.02,r'$2 \ \frac{cm}{s}$',labelpos='W',fontproperties={'weight': 'bold','size':18})
    ax4.grid()
    ax4.set_xlabel('Time (days)')
    handles, labels = ax4.get_legend_handles_labels()
    ax4.legend(handles[::-1], labels[::-1])
    
    f.savefig(savepath + grid  + '_current_analysis_element_' + ("%06d" %(elementmarker[i])) +'.png',dpi=300)
```
